## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SPEND = purchase_data_pd.loc[:,["Gender", "Price", "SN"]]
U_Tot = SPEND.groupby(["SN"]).sum()["Price"].rename("Total Purchase Amount")
U_Ave = SPEND.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
U_Count = SPEND.groupby(["SN"]).count()["Price"].rename("Purchase Count")

user_data = pd.DataFrame({"Purchase Count": U_Count,
                          "Average Purchase Price": U_Ave,
                          "Total Purchase Amount": U_Tot
                          })

user_data.sort_values("Total Purchase Amount", ascending=False).head(5)
```
